```python
import os
import glob
import cv2
from PIL import Image
import shutil
from IPython.display import clear_output

# image_root = "copy_paste/train/images/"
# label_root = "copy_paste/train/labels/"
image_root = "../yolov5/dataset_3/mirror_selfie/images/"
label_root = "../yolov5/dataset_3/mirror_selfie/labels/"

interface_on = True # save interface
save_root = "../yolov5/mirror_refined/"
os.makedirs(os.path.join(save_root, 'images'), exist_ok=True)
os.makedirs(os.path.join(save_root, 'labels'), exist_ok=True)

image_paths = sorted(glob.glob(image_root+"/*.jpg"))
label_paths = sorted(glob.glob(label_root+"/*.txt"))

for path, label_path in zip(image_paths, label_paths):
    try:
        # path = image_paths[0]
        # label_path = label_paths[0]
        print(path, label_path)

        with open(label_path, 'r') as f:
            lines = f.readlines()

        img = cv2.imread(path)
        h, w = img.shape[:2]

        for line in lines:
            l = line.replace('\n', '').split(' ')
            label, xywh = l[0], [float(s) for s in l[1:]]
            # print(xywh)

            x1, y1, x2, y2 = int((xywh[0]-xywh[2]/2)*w), int((xywh[1]-xywh[3]/2)*h), int((xywh[0]+xywh[2]/2)*w), int((xywh[1]+xywh[3]/2)*h)
            # x1,y1,x2,y2 = int(xywh[0]*w),int(xywh[1]*h),int(xywh[0]*w),int(xywh[1]*h)
            # x2, y2 = x1+w1, y1+h1
            if label == '0':
                cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)
            if label == '1':
                cv2.rectangle(img, (x1, y1), (x2, y2), (255,0,0), 2)
    
        # print(path, label_path, h, w, x1,y1,x2,y2)

        display(Image.fromarray(img))
        if interface_on:
            key = input()
            clear_output()
            if key == 's':
                dst = os.path.join(save_root, 'images', os.path.basename(path))
                label_dst = os.path.join(save_root, 'labels', os.path.basename(label_path))
                shutil.copy(path, dst)
                shutil.copy(label_path, label_dst)
            
    except:
        print("except alert")
```
